```python
import pandas as pd
```

```python
# Load the uploaded Excel file
file_path = "/content/yahoo_data.xlsx"
xls = pd.ExcelFile(file_path)

# Display sheet names to understand the data structure
xls.sheet_names
```

```python
# Load the data from the first sheet
df = pd.read_excel(xls, sheet_name="Sheet1")

# Display the first few rows to inspect the dataset
df.head()
```

```python
# Convert 'Date' column to datetime format and sort data by date
df["Date"] = pd.to_datetime(df["Date"])
df = df.sort_values("Date")

# Select relevant columns for forecasting (Date, Close Price)
df = df[["Date", "Close*"]].rename(columns={"Close*": "Close"})

# Check for missing values
df.isnull().sum()
```

```python
from prophet import Prophet

# Prepare data for Prophet
df_prophet = df.reset_index().rename(columns={"Date": "ds", "Close": "y"})

# Initialize and fit Prophet model
prophet_model = Prophet()
prophet_model.fit(df_prophet)

# Create a future dataframe for forecasting (next 365 days)
future = prophet_model.make_future_dataframe(periods=365)
forecast = prophet_model.predict(future)

# Display forecasted data
forecast[["ds", "yhat", "yhat_lower", "yhat_upper"]].tail()
```
