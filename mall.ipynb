```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Load the dataset
file_path = "/content/customer_shopping_data.csv"  # Update the path if needed
df = pd.read_csv(file_path)
```

```python
# Create a new feature for total spending
df['total_spent'] = df['quantity'] * df['price']
```

```python

# Select relevant features for clustering
features = df[['age', 'quantity', 'total_spent']]
```

```python
# Normalize the data
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
```

```python
# Apply K-Means clustering with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df['cluster'] = kmeans.fit_predict(features_scaled)
```

```python

# Display cluster counts
print(df['cluster'].value_counts())
```

```python

# Save the clustered data
df.to_csv("/content/customer_shopping_data.csv", index=False)
```

```python

# Display cluster centers
print("Cluster Centers:")
print(kmeans.cluster_centers_)
```
